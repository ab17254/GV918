# Data wrangling



##  Rename the variables to merge

```
df_brexit.rename({'PCON11CD':'ons_id'}, axis = 1, inplace=True)
```

## Select columns

```
df_brexit.rename({'Figure to use':'leave_pct'}, axis = 1, inplace=True)
```

```
df_brexit_sub = df_brexit[['ons_id', 'leave_pct']]
```

```
df_elec['con_pct'] = df_elec['con'] / df_elec['valid_votes']
df_elec['lab_pct'] = df_elec['lab'] / df_elec['valid_votes']
```

```
df_elec.head()
```

## Merge

```
df_merge = df_elec.merge(df_brexit_sub, on = "ons_id")
```

```
df_merge
```

## Join

```
df_brexit_sub_idx = df_brexit_sub.set_index('ons_id')
df_elec_idx = df_elec.set_index('ons_id')
```

```
df_brexit_sub_idx.head()
```

```
df_elec_idx.head()
```

```
df_elec_idx.join(df_brexit_sub_idx, on = 'ons_id')
```

# Pivoting 

```
df_elec_sub = df_elec[:5][['ons_id', 'constituency_name', 'con', 'lab', 'valid_votes']]
df_elec_sub
```

```
df_melted = df_elec_sub.melt(id_vars = ['ons_id', 'constituency_name', 'valid_votes'],
                 var_name = 'party', value_name = 'vote').sort_values('constituency_name')
```

```
df_melted.pivot(['ons_id', 'constituency_name', 'valid_votes'], 'party').reset_index()
```
